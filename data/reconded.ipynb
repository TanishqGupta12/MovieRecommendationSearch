```python
movie = pd.read_csv('movies.csv' , encoding = "ISO-8859-1")
```

```python
movie.head(1)
```

```python
data=movie[['genres', 'id', 'keywords', 'original_language', 'original_title','overview', 'runtime', 'spoken_languages', 'title' , 'release_date','status','vote_average']]
```

```python
print( data.columns)
data.head(1)
```

```python
data.isna().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isna().sum()
```

```python
def convert (obj):
    genres = []
    for i in ast.literal_eval(obj) :
        genres.append(i['name'])
    return genres
```

```python
 data['genres'] = data['genres'].apply(convert)
```

```python
 data['keywords'] = data['keywords'].apply(convert)
```

```python
data['spoken_languages'][0]
```

```python
def converts(obj):
    objects = []
    count = 0
    for i in ast.literal_eval(obj) :
        if count != 3:
            objects.append(i['name'])
            count =+ 1
        else:
            break
    return objects
```

```python

data['spoken_languages'] = data['spoken_languages'].apply(converts)
```

```python
# data['keywords'].iloc[0] 
```

```python
data['spoken_languages'] = data['spoken_languages'].apply(lambda x:[i. replace(" "," ") for i in x])
data['genres'] = data['genres'].apply(lambda x:[i. replace(" "," ") for i in x])
data['keywords'] = data['keywords'].apply(lambda x:[i. replace(" "," ") for i in x])
```

```python
data.head(5)
```

```python
data['genres'].shape
```

```python
data['genres'] = data['genres'].apply(lambda x : " ".join(x))
data['keywords'] = data['keywords'].apply(lambda x : " ".join(x))  
data['spoken_languages'] = data['spoken_languages'].apply(lambda x : " ".join(x)) 
```

```python
data.head()
```

```python
data['tag'] = data['genres'] +  data['keywords'] 
```

```python
datas = data[['id', 'title', 'genres', 'keywords', 'original_language',
       'original_title', 'overview', 'runtime',
       'tag' , 'release_date','status','vote_average']]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
datas['tag'] = datas['tag'].apply(stem)
```

```python
datas
```

```python
datas['tag'] = data['tag'].apply(lambda x : x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000 , stop_words = 'english')
```

```python
vector = cv.fit_transform( data['tag']).toarray()
```

```python
vector
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity( vector)
```

```python
import pickle 
pickle.dump(similarity ,open('similarity.pkl','wb'))
```

```python
data = sorted(list(enumerate(similarity[0])) ,reverse = True ,key=lambda x: x[1])[1:6]
```

```python
sorted(list(enumerate(similarity[0])) ,reverse = True ,key=lambda x: x[1])[1:6]
```

```python
def recommend (movies):
    movie = datas[datas['title'] == movies ].index[0]
    distance = similarity[ movie]
    movi_list = sorted(list(enumerate( distance)) ,reverse = True ,key=lambda x: x[1])[1:10]
    for  i in  movi_list:
        print(datas.iloc[i[0]].title)
```
